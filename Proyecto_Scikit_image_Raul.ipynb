```python
#Using ray
@ray.remote
def transform_and_save_ray(targetDirectory, df_in, niter):
  from skimage import io
  from skimage.transform import rescale, resize, downscale_local_mean, rotate

  result = []
  for index, elem in df_in.iterrows():
    img_test = elem['image']
    splitted_path = img_test.split("/") 
    img_test_name = splitted_path[len(splitted_path)-1]
    folder_name = elem['label']

    imagen = io.imread(img_test)

    image_resized = resize(imagen, (445, 590),                             
                      anti_aliasing=True)
    target_path = os.path.join(targetDirectory, folder_name)

    for i in range(niter):
      image_rotated = rotate(image_resized, 10*(i+1))

      img_tmp = str(i)+"_"+img_test_name
      path_img = os.path.join(target_path, img_tmp)
      #print(path_img)
      io.imsave(path_img, image_rotated)
      result.append(path_img)

  return result  
```

##Model processing

```python
from sklearn.model_selection import train_test_split
```

```python
dataset_path = Path(path_ray)

file_path = list(dataset_path.glob(r'**/*.png'))
file_path_jpg = list(dataset_path.glob(r'**/*.JPG'))
file_path_jpeg = list(dataset_path.glob(r'**/*.jpeg'))

file_path = [*file_path, *file_path_jpg, *file_path_jpeg]

# create labels from the folder name
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], file_path))
file_path = pd.Series(file_path).astype(str)
labels = pd.Series(labels)

df = pd.concat([file_path, labels], axis=1)

df.columns = ['image', 'label']

df.head()
```

```python
x_train, x_test = train_test_split(df, test_size=0.3,random_state=30)
```

```python
print("Shape of training data", x_train.shape)
print("Shape of test data", x_test.shape)
```

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential, Model
from keras.layers import Dense, Flatten, Dropout, GlobalAveragePooling2D
```

```python
image_data_generator = ImageDataGenerator(rescale = 1./255,
    rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')

train = image_data_generator.flow_from_dataframe(dataframe=x_train, x_col='image', y_col='label', target_size=(445, 590), color_mode='rgb', class_mode='categorical', shuffle=False)
test = image_data_generator.flow_from_dataframe(dataframe=x_test, x_col='image', y_col='label', target_size=(445, 590), color_mode='rgb', class_mode='categorical', shuffle=False)
```

```python
input_shape = (445, 590, 3)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=input_shape ),
    tf.keras.layers.MaxPool2D(pool_size = (2,2)),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size = (2,2)),
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size = (2,2)),
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size = (2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(9, activation='softmax')
])

model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=["accuracy"])

callback = tf.keras.callbacks.EarlyStopping(monitor='accuracy', patience=4)

history = model.fit(train, epochs=10, callbacks=callback)
```
